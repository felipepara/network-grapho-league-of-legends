# Network Study on League of Legends Champions

This notebook use Riot's API data to better understand the interactions between league of legends champions. We will use `networkX` to build our networks and `bokeh` to plot them and visualize data.

What we are going to do is see the most common connections between champions from the same team. In other words, understand what are the most common champions combination in a team.

The dataset we used was gathered from Kaggle:

https://www.kaggle.com/paololol/league-of-legends-ranked-matches

    3) add the ranks from each match, so we can better understand how each rank behaves;

## 1. Libraries & Data 

Below, we import the libraries we are going to use during this study:

```python
import pandas as pd ## As always, to work with data

from IPython.display import Image ## This one we will use to plot images on our notebook.

import networkx as nx ## The lib that creates the network

from bokeh.io import output_file, show ## To show the results
from bokeh.models import (BoxZoomTool, Circle, HoverTool,
                          MultiLine, Plot, Range1d, ResetTool,ColumnDataSource, LabelSet, Slider) # Some bokeh models we will use
from bokeh.models.graphs import from_networkx ## The lib that will connect bokeh and networkX
```

Below, we import the data we will use:

```python
df_matches = pd.read_csv('datasets/matches.csv') ## Data from the players from each match
df_matches.head(5)
```

```python
df_champs = pd.read_csv('datasets/champs.csv') ## Champions Names and Ids
df_champs.head(5)
```

```python
df_class = pd.read_csv('datasets/classes.csv', sep = ';') ## Champions Classification  
df_class.head(5)
```

```python
df_match_data = pd.read_csv('datasets/match_data.csv', sep = ',') ## Data from each match, such as season, duration, and such
df_match_data.head(5)
```

## 2. Data Wrangling

Now, we combine the datasets we have to create the one we are going to use on our model.

We will first combine the dataset that has player information per match with the one that gives us the season the match took place.

```python
df_match_data= df_match_data[['id','seasonid']]
df_matches = pd.merge( df_matches, df_match_data, left_on = 'matchid',right_on = 'id', how = 'left', suffixes = ('','_match'))
df_matches.head(5)
```

Now, we group the champions dataset with their classification.

```python
df_champs= pd.merge(df_champs,df_class,left_on = 'name', right_on = 'champion_name', how = 'left')
df_champs = df_champs.drop('champion_name',axis =1)
```

Here, we define the colors we are going to use for each classification. We will use this when plotting.

```python
colors = pd.DataFrame()
colors['class']=list(df_champs['class'].unique())
colors['color'] = ['red','green','blue','pink','orange','purple','yellow']
df_champs = pd.merge(df_champs,colors, on ='class')
```

```python
df_champs.head(5)
```

Now, we separate the blue team from the red team. And that is a very important step, given that we want to measure the connection between champions from the **same** team.

In our dataset, players 1-5 are from the blue team, players 6-10 are from the red team, and that's how we are going to flag them.

```python
df_matches['team'] = 'BLUE'
df_matches.loc[df_matches['player']>5,'team'] = 'RED'

df_matches['match_team'] = df_matches['matchid'].astype(str) + df_matches['team']  
```

Finally, we merge the matches dataset with the champions dataset, and we have our final table:

```python
df_matches = pd.merge(df_matches,df_champs,left_on = 'championid',right_on='id')
df_matches.head(5)
```

## 3. Creating the Network

Now, we can start building a network with our data.

We will break this process in two main steps: (i) filtering our dataset and (ii) creating the network.

The first step is important because we want to analyze data from the same season. If we analyze our whole dataset without this filter, we will compare champions that were recently added to the game to the ones that were there on previous seasons, and that would create an unbalanced analysis.

Also on the first step, we will create a square matrix that we will use to build our network. The square matrix will present the number of times each champion played alongside with other champions, something like that:

```python
Image(filename = 'images/matrix.PNG')
```

This table will be the very input for our network.

So let's build it:

```python
def update_dataset(season):
    '''
    This funcion is used to create a filtered square matrix that counts each time a champion played alongside
    others.
    
    Parameters
    -Season: The number of the season that will filter our dataset. This dataset have data from season 4-8.
    
    Returns
    -Square Matriz with all the champions
    -Dictionary with the champions and their classification colors
    
    '''
    
    ## Filtering the dataset
    df = df_matches[df_matches['seasonid']==season] 

    ## Creating a dataset that will be used to count the times each champion was in the same team as the others
    df_teams = pd.DataFrame()

    ## We iterate through every champion and count the times it played with the others
    for champ_id in df['name'].unique():
        
        ## We get the list of every match and team the champion participated
        list_of_matches = list(df[df['name']==champ_id]['match_team'].unique())
        df_temp = pd.DataFrame()
        
        ## Now we get all the champions that were in the same team, in the same match.
        df_temp['team_champion'] = df[((df['match_team'].isin(list_of_matches))&(df['name']!=champ_id))]['name']
        
        ## We create a column with the champion name
        df_temp['name'] = champ_id
        
        ## We append it to our dataset
        df_teams = df_teams.append(df_temp) 
    
    ## Now, we group by our dataset by each combination of champions, to see how many times they played in the same team
    df_teams['counter'] = 1
    df_teams = df_teams.groupby(['name','team_champion']).sum().reset_index()
    
    ## We pivot the table, so we can have the square matrix we mentioned before.
    df_cross = df_teams.pivot(index = 'name', columns = 'team_champion', values= 'counter').reset_index()
    df_cross.index = df_cross['name'] 
    df_cross = df_cross.drop('name', axis= 1)
    
    ## Now, we create a dataset with the champions we have on this season.
    df_cham_season  = pd.DataFrame()
    df_cham_season['name'] = list(df['name'].unique())
    df_cham_season = pd.merge(df_cham_season, df_champs[['name','color']], on = 'name', how = 'left') 
    
    ## We set the name and color for each node of our network
    nodes = list(df_cham_season['name'])
    color = list(df_cham_season['color'])
    
    ## Here, we create a dictionary with each champion and its color
    node_color = dict(zip(nodes,color))

    return df_cross, node_color
```

```python
def plot_season_graph(season):
    '''
    This funcion uses the matrix and the dictionary from the previous function and creates our network visualization.
    
    Parameters
    -Season: The number of the season that will filter our dataset. This dataset have data from season 4-8.
    
    Returns
    -Plot: The plot in an HTML file.
    '''
    
    ## We take the outputs from the update_dataset function
    
    df, node_color = update_dataset(season)

    ## Now, we create the graph with networkx
    G = nx.from_numpy_matrix(df.values) 
    
    ## We relabel each node
    G = nx.relabel_nodes(G,dict(enumerate(df.columns)))
    
    ## Setting the color for each node
    nx.set_node_attributes(G,node_color, name = 'node_color')

    ## Creating a Bokeh object
    plot = Plot(plot_width=2000, plot_height=2000,
                x_range=Range1d(-1.1, 1.1), y_range=Range1d(-1.1, 1.1))
    
    ## Setting the title
    plot.title.text = "Interaction for Season {0}".format(season)

    ## Activating hovertool to present the name of the champion
    node_hover_tool = HoverTool(tooltips=[("index", "@index")])
    plot.add_tools(node_hover_tool, BoxZoomTool(), ResetTool())
    
    ## Creating the network graph
    graph_renderer = from_networkx(G, nx.spring_layout, scale=1, center=(0, 0))

    ## Coloring and sizing the nodes
    graph_renderer.node_renderer.glyph = Circle(size=25, fill_color='node_color')
    
    ## Getting the x and y position from each node so we can put a label on them
    pos = graph_renderer.layout_provider.graph_layout 
    x,y = zip(*pos.values())

    ## Creating a ColumnDataSource with the label information, x, y position and what to show
    source = ColumnDataSource({'x':x, 'y':y, 'label': [list(pos.keys())[i] for i in range(len(x))]})
    
    ## Creating the labels and customizing it
    labels = LabelSet(x='x',y= 'y', text = 'label', source = source,
                     text_font_size = '15pt', border_line_color = 'black', 
                      background_fill_color = 'white', text_color = 'black')
    
    ## Adding our graph to the bokeh object
    plot.renderers.append(graph_renderer)
    
    ## Adding the labels to the bokeh object
    plot.renderers.append(labels)

    ## Plotting the graph
    output_file("interactive_graphs.html")
    show(plot)
```

```python
plot_season_graph(8)
    
```
